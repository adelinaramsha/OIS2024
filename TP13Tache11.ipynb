```r
# Charger le jeu de donnes iris
data("iris")

# Dfinir les couleurs pour chaque espce
colors <- c("setosa" = "red", "versicolor" = "green", "virginica" = "blue")

# Dfinir les symboles pour chaque espce
symbols <- c("setosa" = 16, "versicolor" = 17, "virginica" = 18)

# Crer un graphique vide avec les bonnes limites
plot(iris$Petal.Length, iris$Petal.Width, type = "n",
     xlab = "Longueur des ptales (cm)", ylab = "Largeur des ptales (cm)",
     main = "Relation entre la longueur et la largeur des ptales selon l'espce")

# Ajouter des points pour chaque espce
for (species in unique(iris$Species)) {
  species_data <- subset(iris, Species == species)
  points(species_data$Petal.Length, species_data$Petal.Width,
         col = colors[species], pch = symbols[species])
}

# Ajouter une rgression linaire pour chaque espce
for (species in unique(iris$Species)) {
  species_data <- subset(iris, Species == species)
  model <- lm(Petal.Width ~ Petal.Length, data = species_data)
  abline(model, col = colors[species], lwd = 2)
}

# Ajouter une lgende
legend("topleft", legend = unique(iris$Species), col = colors, pch = symbols, title = "Espce")
```
